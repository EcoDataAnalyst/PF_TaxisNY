## analisis de Electric and Alternative Fuel Charging Stations

```python
df = pd.read_csv('../Data/Electric and Alternative Fuel Charging Stations.csv')
df
```

```python
# extraer las estaciones de recarga de solo nueva york
df_ny= df[df['State']== 'NY']
```

```python
df_ny['Fuel Type Code'].unique()
```

```python
df_ny['City'].unique()
#muchas mas ciudades de NY de las necesarias. recortar a solo 'Brooklyn', 'Bronx', 'Queens',
```

```python
# extraccion de las columnas con informacion importante.
df_ny = df_ny[['Fuel Type Code', 'City', 'Status Code', 'Street Address']]
```

```python
df_ny.head(5)
```

```python
# extraccion de la informacion de solo los boroughts Manhattan, Brooklyn, Bronx, Queens,  Staten Island

df_ny = df_ny[ (df_ny['City'] == 'Manhattan') | (df_ny['City'] == 'Brooklyn') 
              | (df_ny['City'] == 'Bronx') | (df_ny['City'] == 'Queens') 
              | (df_ny['City'] == 'Staten Island') ]


df_ny
```

```python
df_ny.groupby('City')['Fuel Type Code'].value_counts().unstack(fill_value=0)
```
