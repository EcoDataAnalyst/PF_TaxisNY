```python
columnas = ["Year", "Manufacturer", "Model", "co2", "co2TailpipeGpm", "fuelType", "eng_dscr", "VClass", "trany"]
```

```python
simply_df = df[columnas].copy()
```

```python
simply_df = simply_df[simply_df["Year"] >= 2017]
```

```python
simply_df = simply_df.drop_duplicates()
```

```python
simply_df
```

```python
df1 = simply_df[(simply_df["Manufacturer"] == "Nissan") & (simply_df["Model"] == "Altima")].drop_duplicates(subset=["Year"])
df2 = simply_df[(simply_df["Manufacturer"] == "Ford") & (simply_df["Model"] == "Fusion FWD")].drop_duplicates(subset=["Year"])
df3 = simply_df[(simply_df["Manufacturer"] == "Ford") & (simply_df["Model"] == "Escape FWD")].drop_duplicates(subset=["Year"])
df4 = simply_df[(simply_df["Manufacturer"] == "Ford") & (simply_df["Model"] == "Taurus FWD")].drop_duplicates(subset=["Year"])
df5 = simply_df[(simply_df["Manufacturer"] == "Ford") & (simply_df["Model"] == "Transit Connect Van FWD")].drop_duplicates(subset=["Year"])
df6 = simply_df[(simply_df["Manufacturer"] == "Chrysler") & (simply_df["Model"] == "Voyager")].drop_duplicates(subset=["Year"])
df7 = simply_df[(simply_df["Manufacturer"] == "Toyota") & (simply_df["Model"] == "Camry")].drop_duplicates(subset=["Year"])
df8 = simply_df[(simply_df["Manufacturer"] == "Toyota") & (simply_df["Model"] == "Prius")].drop_duplicates(subset=["Year"])
df9 = simply_df[(simply_df["Manufacturer"] == "Chevrolet") & (simply_df["Model"] == "Malibu")].drop_duplicates(subset=["Year"])
df10 = simply_df[(simply_df["Manufacturer"] == "Chevrolet") & (simply_df["Model"] == "Impala")].drop_duplicates(subset=["Year"])
df11 = simply_df[(simply_df["Manufacturer"] == "Hyundai") & (simply_df["Model"] == "Sonata")].drop_duplicates(subset=["Year"])
df12 = simply_df[(simply_df["Manufacturer"] == "Lincoln") & (simply_df["Model"] == "MKZ FWD")].drop_duplicates(subset=["Year"])
df13 = simply_df[(simply_df["Manufacturer"] == "Nissan") & (simply_df["Model"] == "NV200 NYC Taxi")].drop_duplicates(subset=["Year"])
df14 = simply_df[(simply_df["Manufacturer"] == "Toyota") & (simply_df["Model"] == "Sienna AWD")].drop_duplicates(subset=["Year"])
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14], axis=0)
```

```python
merged_df
```
